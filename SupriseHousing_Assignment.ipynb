### Surprise Housing Data Analytics Regression Model - Advanced Regression
Problem Statement
Surprise Housing Company, based in the US, aims to leverage data analytics to purchase houses at prices below their actual values and then resell them at higher prices for profit. To achieve this goal, the company seeks to develop a regression model employing regularization techniques to predict the actual value of prospective properties. They have access to a dataset containing sale details of houses in Australia.

The primary objective is to identify significant variables that influence house prices and to assess how effectively these variables explain variations in house prices.

### Solution Approach:
## Data Analysis:
Understand the structure and characteristics of the dataset.
Identify potential variables that could influence house prices.
## Data Cleaning and Feature Selection:
Handle missing values, outliers, and inconsistencies in the dataset.
Select relevant features that are likely to impact house prices based on domain knowledge and statistical analysis.
## Advanced Linear Regression Model:
Implement advanced linear regression techniques to build a predictive model.
Evaluate the model's performance and interpret the coefficients of the selected features.
## Ridge Regression Model:
Apply Ridge regression, a regularization technique, to mitigate overfitting and improve model generalization.
Tune the regularization parameter to find the optimal balance between bias and variance.
## Lasso Regression Model:
Utilize Lasso regression, another regularization method, to perform feature selection by shrinking coefficients towards zero.
Assess the impact of feature selection on model performance and interpret the selected features.
By following these steps, Surprise Housing Company can develop a robust regression model that accurately predicts house prices while identifying significant predictors for informed decision-making in their real estate investments.

#### Identify Missing Values: First, we'll identify which categorical columns have missing values.
#### Imputation Strategy: For each categorical column with missing values, we'll employ an imputation strategy. #### Implementation: We'll apply the chosen imputation strategy to fill the missing values in the categorical columns, ensuring that the integrity of the data is preserved.

```python
def buildLRModel(X_train_param, y_train_param):
    X_train_sm = sm.add_constant(X_train_param)
    
    #create linear regression model using OLS
    lr = sm.OLS(y_train_param, X_train_sm)
    lr_model = lr.fit()
    return lr_model
```

```python
def evaluateVIF (X_train_param):
    vif = pd.DataFrame()
    vif['Features'] = X_train_param.columns
    vif['VIF'] = [variance_inflation_factor(X_train_param.values, i) for i in range(X_train_param.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = 'VIF', ascending = False)
    return vif
```

```python
def get_top_features(model, features, n=5):
    coef = model.coef_
    indices = np.argsort(np.abs(coef))[::-1][:n]
    return features[indices], coef[indices]

top_features, top_coefs = get_top_features(ridge, X_train_rfe.columns[:-1])
print("Top 5 most important predictor variables in the initial ridge model:")
for feature, coef in zip(top_features, top_coefs):
    print(f"{feature}: {coef}")
```

Conversely, the Ridge model excels at lower alpha values, notably at alpha = 0.001. Although the difference is slight, Lasso tends to yield a slightly higher R2 score compared to Ridge in this scenario.
* Neighborhood features such as NoRidge and NridgHt are highly influential in predicting house prices.

* OverallQual (Overall Quality) and OverallCond (Overall Condition) of the house play crucial roles in determining its price.

* Features related to the kitchen (KitchenQual and KitchenAbvGr) and basement (BsmtExposure, BsmtQual, BsmtFinType1, BsmtCond) are important predictors.

* Garage-related features like GarageArea and GarageFinish also contribute to predicting house prices.
